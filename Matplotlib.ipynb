```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Graficos de linhas
np.random.seed(7)
# vetor com numeros inteiros aleatorios entre 1,1500 com 10 amostras

y = np.random.randint(low = 1, high = 1500, size = 10)
```

```python
# insere a primeira linha no plot
plt.plot(y,color = '#749187', marker = '*', ms = 5, mec = 'k', markerfacecolor = 'w', ls = '-.')

#insere a segunda linha no plot
plt.plot(y*2, marker = '+', color = 'k', ms = 5)

#rotulos
plt.xlabel('Eixo X', color = 'red', size = 12)
plt.ylabel('Eixo Y')
plt.title('Meu Primeiro Grafico', color = 'red', loc = 'left')

#gridlines
plt.grid(axis = 'y', color = 'gray',
         linestyle = '--', linewidth = 1,
         alpha = 0.8)
plt.show()
```

```python
# subplots

np.random.seed(6)
x = np.arange(1, 11)
y1 = np.random.randint(1, 400, 10)
y2 = np.random.randint(150, 500, 10)
y3 = np.random.randint(200, 600, 10)

plt.figure(figsize =(15, 5))
plt.suptitle('Figura', fontsize = 15)

plt.subplot(1, 3, 1)
plt.plot(x, y1, color = 'k')
plt.title('Subplot 1', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.subplot(1, 3, 2)
plt.plot(x, y2, color = 'r')
plt.title('Subplot 2', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.subplot(1, 3, 3)
plt.plot(x, y3, color = 'g')
plt.title('Subplot 3', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.tight_layout(pad = 4)
plt.show()


```

```python
fig, ax = plt.subplots(1, 3, figsize = (15, 5))
fig.suptitle('Figura')
ax[0].plot(x, y1, color = 'black')
ax[1].plot(x, y2, color = 'blue')
ax[2].plot(x, y3, color = 'red')

for i in range(3):
      ax[i].set(title = f'Subplot {i+1}', xlabel = 'Eixo X', ylabel = 'Eixo Y')
```

```python
#Dados economicos
import plotly.express as px
```

```python
#dataset gapminder
px.data.gapminder()
```

```python
# atribui o dataset a variavel df e altera o index para coluna year
df = px.data.gapminder().query('country == "Brazil"').set_index('year')
```

```python
df.head()
```

```python
plt.plot(df.index, df['gdpPercap'])
plt.title('PIB per capita do Brasil')
plt.ylabel('PIB per capita')
plt.xlabel('Tempo');
```

```python
df.head()
```

```python
title = 'Relacao entre expectativa de vida e renda per capita no Brasil'

plt.figure(figsize = (12, 4))
plt.scatter(df['lifeExp'], df['gdpPercap'], cmap = 'viridis')
plt.xlabel('Expectativa de Vida')
plt.ylabel('Renda per capita')
plt.title(title, loc = 'left')
plt.show()
```

```python
#grafico de barra
plt.bar(x = df.index,
        height = df['pop'],
        color = 'red'
        )
plt.title('Populacao brasileira');
```

```python
# def = criar uma funcao. NOme_da_Funcao = filtrar_continente
def filtrar_continente(continente):
    df = px.data.gapminder()
    df = df[df['continent'] == continente]
    return df
```

```python
def filtrar_pais(pais, variavel):
    'Filtrar algum pais da amostra para as variaveis "pop", "gdpPercap" e "lifeExp"'
    df = px.data.gapminder()
    df = df[df['country'] == pais][variavel]
    return df
```

```python
americas = filtrar_continente('Americas')
paises = americas['country'].unique()
plt.figure(figsize = (12, 8))
for pais in paises:
    plt.scatter(filtrar_pais(pais = pais,
                             variavel = 'lifeExp'
                             ),
                filtrar_pais(pais = pais,
                             variavel = 'gdpPercap'
                             ))
plt.legend(labels = paises, loc = 'best' )
plt.title('Relacao entre renda per capita e expectativa de vida', loc = 'left')
plt.xlabel('Expectativa de Vida')
plt.ylabel('Renda per capita');   
```
